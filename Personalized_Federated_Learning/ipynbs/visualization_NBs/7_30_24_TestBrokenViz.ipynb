```python
import numpy as np
import h5py
import torch
import os
import matplotlib.pyplot as plt
```

```python
general_results_path = r"C:\Users\kdmen\Desktop\Research\personalization-privacy-risk\Personalized_Federated_Learning\results"
fedavg_filename = "FedAvg_test.h5"

directory_str = "07-30_12-09_FedAvg_LinRegr100_kfold1"
h5_path = os.path.join(general_results_path, directory_str, fedavg_filename)
with h5py.File(h5_path, 'r') as f:
    a_group_key = list(f.keys())
    print(a_group_key)
    fedavg_train_loss = f['rs_train_loss'][()]
    fedavg_test_loss = f['rs_test_loss'][()]
    
    fedavg_client_testing_logs = f['client_testing_logs']
    fedavg_cost_func_tuples_by_client = f['cost_func_tuples_by_client']
    fedavg_gradient_norm_by_client = f['gradient_norm_lists_by_client']

    client_testing_logs_dict = {}
    for key in fedavg_client_testing_logs:
        client_testing_logs_dict[key] = fedavg_client_testing_logs[key][()]

    fedavg_cost_func_tuples_dict = {}
    for key in fedavg_cost_func_tuples_by_client:
        fedavg_cost_func_tuples_dict[key] = fedavg_cost_func_tuples_by_client[key][()]

    fedavg_gradient_norm_dict = {}
    for key in fedavg_gradient_norm_by_client:
        fedavg_gradient_norm_dict[key] = fedavg_gradient_norm_by_client[key][()]
```
