## Second Batch of Runs (After Model Saving Added)

```python
cv_results_path = r"C:\Users\kdmen\Desktop\Research\personalization-privacy-risk\PythonVersion\PythonSimsRevamp\results"
directory_str = "08-13_10-54_PFAFO_GDLS"
filename = "GDLS_PFAFO_GDLS_CrossValResults.h5" 
h5_path = os.path.join(cv_results_path, directory_str, filename)
with h5py.File(h5_path, 'r') as f:
    a_group_key = list(f.keys())#[0]
    PFA_AvgCV_val_loss = f['AveragedCV_local_test_error_log'][()]
    PFA_AvgCV_train_loss = f['AveragedCV_local_train_error_log'][()]

cv_results_path = r"C:\Users\kdmen\Desktop\Research\personalization-privacy-risk\PythonVersion\PythonSimsRevamp\results"
directory_str = "08-13_10-52_FEDAVG"
filename = "GDLS_FEDAVG_CrossValResults.h5"
h5_path = os.path.join(cv_results_path, directory_str, filename)
with h5py.File(h5_path, 'r') as f:
    a_group_key = list(f.keys())#[0]
    FedAvg_AvgCV_val_loss = f['AveragedCV_local_test_error_log'][()]
    FedAvg_AvgCV_train_loss = f['AveragedCV_local_train_error_log'][()]
```

```python
plt.plot(PFA_AvgCV_val_loss, label="PFA")
plt.plot(FedAvg_AvgCV_val_loss, label="FedAvg")
plt.legend()
plt.xlabel("Training Round")
plt.ylabel("Cost")
plt.title("Test Loss During Training")
plt.show()
```

```python
# Combine the data into a list
data = [PFA_AvgCV_val_loss, FedAvg_AvgCV_val_loss]

# Create a box plot
plt.figure(figsize=(12, 8))
plt.boxplot(data, patch_artist=True)#, vert=False)
plt.title('CPHS Time Domain Validation Error: Norm of p_ref - p_act')
plt.ylabel('Norm')
plt.xlabel(("Experimental Trial"))
plt.xticks([1, 2], ["PFA_AvgCV_val_loss", "FedAvg_AvgCV_val_loss"])
plt.show()
```
